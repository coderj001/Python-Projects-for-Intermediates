```python
def read_data(file):
    file = open(file,"r")
    data = file.readlines()
    file.close()
    ids = []
    for line in data:
        if line == "\n" or line == " ":
            pass
        else:
            ids.append(line.split()[0])
            #print(float(line.split()[0]))
    #print(ids)
    return ids
#read_data("customer.txt")
```

```python
def add_to_file(customer_id,salary, age):
    file = open("customer.txt","r")
    data = file.readlines()
    file.close()
    data.append(customer_id +"    "+salary +"    " + age + "\n")
    #print(data)
    file = open("customer.txt","w")
    for Customer in data:
        file.write(Customer)
    file.close()
    print("Customer Successfully Added") 
```

```python
def new_customer():
    customer_id = input("Enter Customer ID: ")
    if customer_id.isdigit() == False:
        print("invalid Customer Id")
        new_customer()
        
    salary = input("Enter Salary:")
    
    age = input("Enter Age: ")
    
    file = "customer.txt"
    ids = read_data(file)
    #print(ids)
    if customer_id in ids:
        print("Customer Id already exist")
    else:
        add_to_file(customer_id,salary, age)
    
    
```

```python
def compute_monthly_payment_amount():
    customer_id = input("Enter Customer ID: ")
    loan = float(input("Loan: "))
    annual_rate = float(input("Annual Rate: "))
    years = float(input("Years: "))
    num_payments = years*12
    monthly_interest = annual_rate/12
    term = pow((1 + monthly_interest),num_payments)
    monthly_payment = (loan * monthly_interest * term) / (term - 1)
    print("Monthly payment: SR {}".format(round(monthly_payment,2)))

    
#Read loan, annual rate, years
#Num of payments is equal to years multiplied by 12
#Monthly interest rate is equal to anual rate divided by 12
#Term is equal to  (1 + monthly interest rate)numPayments
#Monthly payment is equal to  (loan multiplied by monthly interest rate multiplied by term)  divided by (term – 1)
#Display the monthly payment
#Balance is equal to loan
#For each month of the loan
#•	Monthly interest is equal to monthly interest rate multiplied by balance
#•	Principal is equal to monthly interst subctracted from monthly payment if it’s not the final month. Otherwise, princaple is equal to balance, and monthly payemt is equal to balance plus monthly interest.
#•	Balance is equal to principal subctracted from the balanc
#•	Display month, monthInterest, principal, balance
```

```python
def save_data(data,customer_id):
    lines = data
    customer_id = customer_id
    file_name = "{}_amortization_table.txt".format(customer_id) 
    file = open(file_name,"w")
    for data in lines:
        file.write(data)
    file.close()
    print("Amortization table  is saved as {}".format(file_name))
    
    
    
```

```python
def display_amortization_table():
    customer_id = input("Enter Customer ID: ")
    loan = float(input("Loan: "))
    annual_rate = float(input("Annual Rate: "))
    years = int(input("Years: "))
    num_payments = years*12
    monthly_rate = annual_rate/12
    term = pow((1 + monthly_rate),num_payments)
    monthly_payment = (loan * monthly_rate * term) / (term - 1)
    balance = loan
    monthly_interest = monthly_rate * balance 
    principal = monthly_payment - monthly_interest
    print("Month Interest Principal Balance\n-------------------------------------------")
    d1 = str(1)+"    "+str(round(monthly_interest,2))+"    "+str(round(principal,2))+"     "+str(round(balance-principal,2)) + "\n"
    
    print(d1)
    
    
    data = ["Month Interest Principal Balance\n-------------------------------------------\n",d1]
    
    for month in range(2,num_payments+1):
        balance -= principal 
        monthly_rate = annual_rate/12
        term = pow((1 + monthly_rate),num_payments)
        monthly_payment = (loan * monthly_rate * term) / (term - 1)

        monthly_interest = monthly_rate * balance 
        principal = monthly_payment - monthly_interest
        d2 = str(month)+"    "+ str(round(monthly_interest,2))+"    "+str(round(principal,2))+"     "+str(abs(round(balance-principal,2))) + "\n"
        print(d2)        
        data.append(d2)
        
    save_data(data,customer_id)

#'''Prompt for and read the customerID, loan, rate, and years.
#    Then display an amortization table that 
#    shows month, interest amount, principal, and balance.
#    (Suppose that the customer is exists and eligible)
```
