```python
def check(Password_Or_Pin):
    nums = [str(num) for num in range(10)]
    digitcount = []
    for var in Password_Or_Pin:
        if var not in nums:
            return True
        digitcount.append(Password_Or_Pin.count(var)) 
    check_count =  [VarInDigitcount for VarInDigitcount in digitcount if int(VarInDigitcount) != 1]  
    if len(check_count)>0:
        return True
    else:
        return False
```

```python
def changepin1():
    new_pin = input("Enter new 4 Digit pin (should be unique 4-digits) : >")
    if len(new_pin)==4:
        while check(new_pin):
            return changepin1()
        else:
            return new_pin
    else:
        return changepin1()
```

```python
def game():
    
    print("H A N G M A N")
    print("             A game of words")
    print("-----------------------------------")
    words = ["python", "java", "kotlin", "javascript"]
    random.seed()
    while True:
        menu_selection = ""
        while menu_selection not in ["1", "2"]:
            menu_selection = input('Type "1" to play the game, "2" to go back to menu ')
        if menu_selection == "2":
            menu()
            break

        word = random.choice(words)
        word_length = len(word)
        hint = ["-"] * word_length
        lives = 8
        previous_guesses = []
        while lives > 0:
            print()
            print("".join(hint))
            guess = input("Input a letter: ")
            if len(guess) != 1:
                print("You should input a single letter")
                continue
            if not (guess.isascii() and guess.islower()):
                print("It is not an ASCII lowercase letter")
                continue
            if guess in previous_guesses:
                print("You already typed this letter")
                continue
            previous_guesses.append(guess)
            if guess in word:
                for i in range(word_length):
                    if word[i] == guess:
                        hint[i] = guess
                if "".join(hint) == word:
                    print("You guessed the word!")
                    print("You survived!")
                    break
            else:
                print("No such letter in the word")
                lives -= 1

        if lives == 0:
            print("You are hanged!")

        print()

    
```
